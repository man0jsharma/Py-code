```
try:
    # For Python 3.0 and later
    from urllib.request import urlopen
except ImportError:
    # Fall back to Python 2's urllib2
    from urllib2 import urlopen
from bs4 import BeautifulSoup

import requests
import urllib2
```

```
page = requests.get("https://www.amazon.com/s/ref=nb_sb_noss?url=search-alias%3Daps&field-keywords=thermometer&rh=i%3Aaps%2Ck%3Athermometer")
html_contents = page.text

print(html_contents)


```

```
# import urllib2
from urllib.request import build_opener
opener = build_opener()
opener.addheaders = [('User-agent', 'Mozilla/5.0')]
response = opener.open('https://www.amazon.com/s/ref=nb_sb_noss?url=search-alias%3Daps&field-keywords=thermometer&rh=i%3Aaps%2Ck%3Athermometer')
html_contents = response.read()

print(html_contents)
```
